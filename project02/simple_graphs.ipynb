```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
x = [0.01, 0.05, 0.25, 0.5, 1]

y_lr_dirty = [0.796, 0.799, 0.800, 0.800, 0.799]
y_lr_clean = [0.803, 0.805, 0.804, 0.803, 0.804]
y_lr_no_du = [0.796, 0.797, 0.797, 0.797, 0.797]
y_lr_stemm = [0.803, 0.805, 0.804, 0.803, 0.804]
y_lr_lemma = [0.803, 0.805, 0.804, 0.803, 0.804]
y_lr_st_wo = [0.796, 0.798, 0.800, 0.800, 0.799]

fig, ax = plt.subplots()
ax.plot(x, y_lr_dirty, marker="o", linestyle='--', label="raw")
ax.plot(x, y_lr_clean, marker="o", linestyle='-',  label="clean")
ax.plot(x, y_lr_no_du, marker="o", linestyle='--', label="no duplicates")
ax.plot(x, y_lr_stemm, marker="o", linestyle='--', label="normalized stemming")
ax.plot(x, y_lr_lemma, marker="o", linestyle=':',  label="normalized lemmatization")
ax.plot(x, y_lr_st_wo, marker="o", linestyle='--', label="removed stop words")

ax.set(xlabel='regularization strength', ylabel='accuracy',
       title='logistic regression')
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

fig.savefig("plots/logreg.png", dpi=300)
plt.show()
```

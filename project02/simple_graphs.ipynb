```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
x = [0.01, 0.05, 0.25, 0.5, 1]

y_lr_dirty = [0.796, 0.799, 0.800, 0.800, 0.799]
y_lr_clean = [0.803, 0.805, 0.804, 0.803, 0.804]
y_lr_no_du = [0.796, 0.797, 0.797, 0.797, 0.797]
y_lr_stemm = [0.803, 0.805, 0.804, 0.803, 0.804]
y_lr_lemma = [0.803, 0.805, 0.804, 0.803, 0.804]
y_lr_st_wo = [0.796, 0.798, 0.800, 0.800, 0.799]

fig, ax = plt.subplots()
ax.plot(x, y_lr_dirty, marker="o", linestyle='-',  label="raw")
ax.plot(x, y_lr_clean, marker="o", linestyle='-',  label="clean")
ax.plot(x, y_lr_no_du, marker="o", linestyle='-',  label="no duplicates")
ax.plot(x, y_lr_stemm, marker="o", linestyle='--', label="normalized stemming")
ax.plot(x, y_lr_lemma, marker="o", linestyle=':',  label="normalized lemmatization")
ax.plot(x, y_lr_st_wo, marker="o", linestyle='--',  label="removed stop words")

ax.set(xlabel='regularization strength', ylabel='accuracy',
       title='logistic regression')
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

fig.savefig("plots/logreg.png", dpi=300)
plt.show()
```

```python
x = [0.01, 0.05, 0.25, 0.5, 1]

y_lr_dirty = [0.818, 0.822, 0.821, 0.819, 0.812]
y_lr_clean = [0.820, 0.823, 0.822, 0.820, 0.811]
y_lr_no_du = [0.811, 0.813, 0.812, 0.809, 0.799]
y_lr_stemm = [0.805, 0.808, 0.807, 0.805, 0.795]
y_lr_lemma = [0.809, 0.811, 0.810, 0.808, 0.797]
y_lr_st_wo = [0.808, 0.810, 0.809, 0.807, 0.801]

fig, ax = plt.subplots()
ax.plot(x, y_lr_dirty, marker="o", linestyle='-',  label="raw")
ax.plot(x, y_lr_clean, marker="o", linestyle='--', label="clean")
ax.plot(x, y_lr_no_du, marker="o", linestyle='-',  label="no duplicates")
ax.plot(x, y_lr_stemm, marker="o", linestyle='-',  label="normalized stemming")
ax.plot(x, y_lr_lemma, marker="o", linestyle='-',  label="normalized lemmatization")
ax.plot(x, y_lr_st_wo, marker="o", linestyle='-',  label="removed stop words")

ax.set(xlabel='regularization strength', ylabel='accuracy',
       title='SVM')
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

fig.savefig("plots/svm.png", dpi=300)
plt.show()
```

```python
x = [0.01, 0.05, 0.25, 0.5, 1]

y_lr_simpl = [0.819, 0.823, 0.823, 0.821, 0.818]
y_lr_2gram = [0.851, 0.855, 0.849, 0.844, 0.839]
y_lr_3gram = [0.857, 0.861, 0.857, 0.854, 0.850]
y_lr_4gram = [0.858, 0.861, 0.857, 0.855, 0.853]
y_lr_tfidf = [0.858, 0.861, 0.857, 0.855, 0.852]

fig, ax = plt.subplots()
ax.plot(x, y_lr_simpl, marker="o", linestyle='-', label="simple")
ax.plot(x, y_lr_2gram, marker="o", linestyle='-',  label="2-grams")
ax.plot(x, y_lr_3gram, marker="o", linestyle='-', label="3-grams")
ax.plot(x, y_lr_4gram, marker="o", linestyle='--', label="4-grams")
ax.plot(x, y_lr_tfidf, marker="o", linestyle=':',  label="TF-IDF")

ax.set(xlabel='regularization strength', ylabel='accuracy',
       title='Improved SVM')
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

fig.savefig("plots/improved_svm.png", dpi=300)
plt.show()
```
